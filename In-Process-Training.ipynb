```python
%matplotlib notebook
import tensorwatch as tw

# train = tw.WatcherClient(port=0)
train = tw.WatcherClient()

def show(y,x_name,y_name,event_name,wc=train):
    loss_stream = train.create_stream(expr='lambda d:{}'.format(y), event_name=event_name)
    loss_plot = tw.Visualizer(loss_stream, vis_type='line', xtitle=x_name, ytitle=y_name)
    loss_plot.show()
```
